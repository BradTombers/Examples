https://www.hockey-reference.com/play-index/goal_finder.cgi?request=1&match=goallist&grouping=p&team_id=FLA&season_start=1&season_end=-1&rookie=N&age_min=0&age_max=99&is_playoffs=N&game_num_type=team&game_month=0&game_type%5B%5D=R&game_type%5B%5D=OT&game_type%5B%5D=SO&period%5B%5D=4&order_by=date_game

put in csv file 

```python
import pandas as pd 
import numpy as np 
import datetime as dt 

df = pd.read_csv('panthers.csv')
df.index = pd.to_datetime(df.Date)

df.head()
```

```python
df['season_year'] = np.where(df.index.month <= 6, df.index.year - 1, df.index.year)
df_new = df[['Scorer','Assist1','Assist2','season_year']]
df_new = pd.melt(df_new, id_vars=['season_year'], value_vars=['Scorer', 'Assist1','Assist2'])
```

```python
df_new.head()
```

```python
df_new = df_new.sort_values('season_year',ascending = 0)
```

```python
final = df_new.groupby(['season_year','value']).size()
```

```python
final = final.sort_values()
```
